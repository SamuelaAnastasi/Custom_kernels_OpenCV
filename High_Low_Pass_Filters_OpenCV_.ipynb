# Creating a Filter, Edge Detection

### Import resources and display image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline

# Read in the image
image_lane = mpimg.imread('images/curved_lane.jpg')

plt.imshow(image_lane)
```

### Convert the image to grayscale

```python
# Convert to grayscale for filtering
gray_lane = cv2.cvtColor(image_lane, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_lane, cmap='gray')
```

###TODO: Create a custom kernel
The Sobel filter is very commonly used in edge detection and in finding patterns in intensity in an image. Applying a Sobel filter to an image is a way of taking (an approximation) of the derivative of the image in the x or y direction, separately.

```python
# Create a custom kernel
## TODO: Create and apply a Sobel x operator
sobel_x_3 = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])


# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
image_sobel_x_3 = cv2.filter2D(gray_lane, -1, sobel_x_3)

plt.imshow(image_sobel_x_3, cmap='gray')
```

```python
# Create a custom kernel
# 3x3 array for edge detection 
# Create and apply a Sobel y operator
sobel_y_3 = np.array([[ -1, 0, 1], 
                   [ -2, 0, 2], 
                   [ -1, 0, 1]])


# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
image_sobel_y_3 = cv2.filter2D(gray_lane, -1, sobel_y_3)

plt.imshow(image_sobel_y_3, cmap='gray')
```

```python
# Load new image
# Read in the image
image_lines = mpimg.imread('images/white_lines.jpg')

plt.imshow(image_lines)
```

```python
# Convert to grayscale for filtering
gray_lines = cv2.cvtColor(image_lines, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_lines, cmap='gray')
```

```python
# blurr image with OpenCV
# average filter
avg_filter = cv2.blur(gray_lines,(50,50)) # kernel 50x50

plt.imshow(avg_filter, cmap='gray')
```

```python
# blurr image with OpenCV
# Gaussian filter
gaussian_blur_filter = cv2.GaussianBlur(gray_lines, (55,55),0)

plt.imshow(gaussian_blur_filter, cmap='gray')
```

```python
# blurr image with OpenCV
# Median filter
median_blur_filter = cv2.medianBlur(gray_lines,55)

plt.imshow(median_blur_filter, cmap='gray')
```

```python
# blurr image with OpenCV
# Bilateral filter
bilateral_filter = cv2.bilateralFilter(gray_lines,35,125,125) 

plt.imshow(bilateral_filter, cmap='gray')
```
